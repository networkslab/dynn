```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import os
import matplotlib
import matplotlib.lines as mlines

```

```python
file_name = 'acc.csv'
df = pd.read_csv(file_name)
sns.set(font_scale=1.5)  # crazy big


acc_g = []
gated_acc_g = []
for g in range(6):
   
    acc = list(df[df['key']=='train/acc'+str(g)]['value'])[-1]
    gated_acc = list(df[df['key']=='train/gated_acc_'+str(g)]['value'])[-1]
    acc_g.append(acc)
    gated_acc_g.append(gated_acc)
plt.xlabel('gates')
plt.ylabel('accuracy')

plt.plot(acc_g,'o', label=r'acc over $\mathcal{D}$')
plt.plot(gated_acc_g,'o', label=r'acc over gated $\mathcal{D}^l$')

plt.xticks(range(6), [r"$p_{\theta}^"+str(i)+"$" for i in range(6)])


plt.legend()
plt.tight_layout()
plt.savefig('gated_vs_all_acc.pdf')
```
